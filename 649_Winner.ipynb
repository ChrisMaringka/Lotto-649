# Import Libraries

```python
import requests
import pandas as pd
import time
import datetime
import numpy as np
from matplotlib import pyplot as plt
from bs4 import BeautifulSoup
from selenium import webdriver
```

# Defining rules

```python
# Start and end date of the range to scrape
start_date = datetime.date(1988, 8, 24)
end_date = datetime.date(2023, 2, 25)

# Define the delta (in days)
delta = datetime.timedelta(days=1)

# Adding chrome_options

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--incognito")
driver = webdriver.Chrome(options=chrome_options)
```

#  Looping the URLs and Scraping the Winning Numbers

```python
# Define the base URL
base_url = "https://www.playnow.com/lottery/lotto-649-winning-numbers/?date="

# Loop through the date range and generate the URLs
with open('urls.txt', 'a') as file:
    current_date = start_date
    while current_date <= end_date:
        if current_date.weekday() in [2, 5]:
            url = base_url + current_date.strftime('%d/%m/%Y')
            file.write(url + '\n')
            print(url)
        current_date += delta
```

```python
# Read the URLs from urls.txt
with open('urls.txt', 'r') as file:
    urls = file.read().splitlines()

lottoNumbers = '.product-winning-numbers__number_six49'

driver = webdriver.Chrome(options=chrome_options)

# Initialize an empty list to store the data
data = []

for url in urls:
    driver.get(url)
    time.sleep(2)
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Extract the data from the HTML using BeautifulSoup
    numbers = [number.text.strip() for number in soup.select(lottoNumbers)]

    # Add the data to the list
    data.append(numbers)

# Create a DataFrame from the list of lists
df = pd.DataFrame(data, columns=['No1', 'No2', 'No3', 'No4', 'No5', 'No6'])
```

```python
df = df.dropna(how='any',axis=0)
```

```python
df = df.astype({"No1":'int', "No2":'int',"No3":'int',"No4":'int',"No5":'int',"No6":'int'})
```

Following function is basically to ensure that number is formatted in two-digits instead of one-digit (e.g: 05 instead of 5). That way when the app is running, the combination is properly formatted.

```python
concat_digits = lambda x: f"{x['No1']:02d}{x['No2']:02d}{x['No3']:02d}{x['No4']:02d}{x['No5']:02d}{x['No6']:02d}"

df['com'] = df.apply(concat_digits, axis=1)
```

```python
user_input = input("Enter your six-numbers combination in 112233445566 format, starting from the smallest number to the largest. For single-digit  number, please apply '0' in front of it: ")

if df['com'].isin([user_input]).any():
    print("This number has been won before")
else:
    print("This combination has not been won yet. Go for it!")
```
